```python
import streamlit as st
import speech_recognition as sr
```

#**1. Select API**

```python
opt = st.selectbox('Select API', ['amazon', 'google', 'azure', 'ibm'], placeholder='choose one API')
```

#**2. Improve the error handling in the *transcribing_speech()***

```python
def transcribing_speech(lang, opt):
  r = sr.Recognizer()
  with sr.Microphone() as source:
    st.info('speak')
    audio = r.listen()
    if st.button('transcribe'):
      st.write('transcribing ....')
    try:
      if opt == 'azure':
        text = r.recognize_azure(audio, language=lang)
      elif opt == 'amazon':
        text = r.recognize_amazon(audio)
      elif opt=='google':
        text = r.recognize_google(audio, language=lang)
      elif opt == 'ibm':
        text = r.recognize_ibm(audio, language=lang)

      return text
    except:
      return 'there are some mistakes'
```

#**3. Save transcribed text to a file**

```python
def save(text):
  text = transcribing_speech()
  name_file = st.text_input('enter the name of file')
  f = open('text.txt', 'w+')
  f.write(text)

  st.download_button(label='download', data=f, file_name=name_file+'txt')
```

#**4. Choose the language**

```python
option_lang = st.selectbox('choose a language', ['en-US', 'fr-FR'])
```

#**5. Feature to pause and resume speech recognition process**

```python
listen = True
pause = False
resume_btn = st.button('resume')
pause_btn = st.button('pause')
quit_btn = st.button('quit')
while listen:
  if not pause_btn:
    transcribing_speech(option_lang, opt)
  else:
    if resume_btn:
      pause_btn = False
    if quit_btn:
      listen = False
```
