#**2. Clean the dataset**
* There are no missing value.
* So we will only transform object value into numerical value

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# for the 'job', 'marital' and 'contact' column
encoder = LabelEncoder()
for feat in ['job', 'marital', 'contact']:
  df[feat] = encoder.fit_transform(df[feat])
```

```python
# for the 'education' column
df['education'].replace(['unknown', 'primary', 'secondary', 'tertiary'], [0, 1, 2, 3], inplace=True)
```

```python
# for columns that have 'yes' and 'no' values
df.replace(['yes', 'no'], [1, 0], inplace=True)
```

```python
# for 'poutcome' column
df['poutcome'].replace(['failure', 'success', 'other', 'unknown'], [0, 1, 2, 3], inplace=True)
```

```python
# for month column
month_val = np.unique(df['month'])
month_val = list(month_val)
print(month_val)
```

```python
df['month'].replace(month_val, [4, 8, 12, 2, 1, 7, 6, 3, 5, 11, 10, 9], inplace=True)
```

```python
df.info()
```

#**3. Build a  deep learning network to predict the marketing campaign outcome**

```python
features = [i for i in df.columns if i!='deposit']
target = 'deposit'
```

```python
x = df[features]
y = df[target]
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
model = Sequential([
    Dense(units=60, activation='relu'),
    Dense(units=12, activation='relu'),
    Dense(units=5, activation='relu'),
    Dense(units=1, activation='sigmoid')
])
```

```python
model.compile(optimizer=tf.optimizers.Adam(learning_rate=0.0005), loss=tf.keras.losses.BinaryCrossentropy(), metrics='accuracy')
```

```python
model.fit(x, y, epochs=200, verbose=2)
```

```python
y_pred_prob = model.predict(x)
```

```python
print(y_pred_prob)
```

```python
y_pred_label = y_pred_prob.round().astype('int')
```
