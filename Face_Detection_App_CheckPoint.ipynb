#**4. Adjust minNeighbors parameter**

```python
st.markdown('slider the point to choose the value of min neighbor')
min_neighbor = st.slider('Adjust the minNeighbors ', 1, 5, 20)
```

#**5. Adjust the scaleFactor parameter**

```python
st.markdown('slide the point to choose the value of scale factor')
scaleFactor = st.slider('Adjust scale factor ', 1.0, 0.1, 10.0)
```
