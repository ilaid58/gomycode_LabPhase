```python
import pandas as pd
```

```python
data = {'Name': ['John', 'Mary', 'Bob', 'Sarah', 'Tom', 'Lisa'],
        'Department': ['IT', 'Marketing', 'Sales', 'IT', 'Finance', 'Marketing'],
        'Age': [30, 40, 25, 35, 45, 28],
        'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
        'Salary': [50000, 60000, 45000, 55000, 70000, 55000],
        'Experience': [3, 7, 2, 5, 10, 4]}

employee_df = pd.DataFrame(data)
employee_df
```

1. use **iloc** to select the first three rows

```python
employee_df.iloc[:3]
```


2. Use **loc** to select all row where Department is *Marketing*

```python
employee_df.index = employee_df['Department']
employee_df.loc['Marketing']
```


3. Use **iloc** to select the Age and Gender columns for the first 4 rows

```python
employee_df.index = range(employee_df.shape[0])
employee_df[['Age', 'Gender']].iloc[:4]
```

4. Use **iloc** to select the Salary and Experience columns where Gender is 'Male'

```python
employee_df.index = employee_df['Gender']
employee_df[['Salary', 'Experience']].loc['Male']
```
