## 2.1) Replace categorical value by numerical value

```python
feature_object = [i for i in df.columns if df[i].dtypes == object]
```

```python
feature_object
```

```python
label = LabelEncoder()
for col in feature_object:
  df[col] = label.fit_transform(df[col])
```

```python
df
```

## 2.2) Fill NaN value

'Year' (int) and 'Publisher' (int) columns have missing values


```python
df['Year'].fillna(df['Year'].mode()[0], inplace=True)
```

```python
df['Publisher'].fillna(df['Publisher'].mode()[0], inplace=True)
```

# 3: Area Chart to show trend of global sales over years

```python
!pip install plotly==5.13.0
```

```python
import plotly.express as px
import plotly.graph_objs as go
import plotly.offline as pyo
```

```python
fig = px.area(df, x='Global_Sales', y='Year', range_y=[df['Year'].min(), df['Year'].max()])
fig.show()
```

# 4: Histogram to show distribution of global sales

```python
fig = px.histogram(df, x='Global_Sales', nbins=50)
fig.show()
```

# 5: Scatter plot to show the relationship between North American sales and European sales

```python
trace = go.Scatter(x=df['NA_Sales'], y=df['EU_Sales'], mode='markers')
data = [trace]
layout = go.Layout(title='Noth American sales and European sales',
                   xaxis=dict(title='Year of Sale'),
                   yaxis=dict(title='Price'))
fig = go.Figure(data=data, layout=layout)
pyo.plot(fig)
```

```python
# open the html file
from IPython.core.display import display, HTML
display(HTML('temp-plot.html'))
```

# 6: Scatter plot to show the relationship between Japanese sales and Other sales

```python
trace = go.Scatter(x=df['JP_Sales'], y=df['Other_Sales'], mode='markers')
data = [trace]
layout = go.Layout(title='Japanese and Othe Sales',
                   xaxis = dict(title='japanese sales'),
                   yaxis = dict(title='Other sales'))
fig = go.Figure(data=data, layout=layout)
pyo.plot(fig)
```
