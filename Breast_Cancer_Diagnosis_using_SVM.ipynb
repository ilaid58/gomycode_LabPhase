#**1. Import the necessary libraries**

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.svm import *
```

#**2. Load the breast dataset**

```python
data = load_breast_cancer()
```

#**3. Assigning the features and target**

```python
df_features = data.data
df_target = data.target
print('features data : ', df_features)
print('\ntarget data : ', df_target)
```

#**4. Split the data into train and test set (with 30% test set)**

```python
x_train, x_test, y_train, y_test = train_test_split(df_features, df_target, test_size=0.3)
```

#**5. Create a model SVM**
We have a classification dataset so we will use svc model

```python
model = SVC()
```

#**6. Fit the model to training set**

```python
model.fit(x_train, y_train)
```

#**7. Predict target variable using test set**

```python
y_pred = model.predict(x_test)
```

#**8. Calculate accuracy and confision matrix**

```python
accur_score = accuracy_score(y_test, y_pred)
confus_matrix = confusion_matrix(y_test, y_pred)
```

#**9. print accuracy and accuracy**

```python
print('accuracy score : ', accur_score)
print('\nTP : ', confus_matrix[0, 0])
print('TN : ', confus_matrix[1, 1])
print('FP : ', confus_matrix[0, 1])
print('FN : ', confus_matrix[1, 0])
```
